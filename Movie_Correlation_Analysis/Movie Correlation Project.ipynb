```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the Configurationn of the Plot we will create


# Read in the Data

df = pd.read_csv(r'C:\Users\PMY\Downloads\movies.csv')
```

```python
# Lets look at the data

df.head()
```

```python
# lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# dtaa type for or columns

df.dtypes
```

```python
# change data type of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#create correct year colunns

df['yearcorrect'] = df['released'].astype(str).str[:4]

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```
